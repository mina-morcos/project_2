```python
#enter region name and number of recipes here
origin = "latin american"
number_of_recipe = "10"
rapid_api_key = ""
```

```python
#Get recipes id base on region
url = "https://spoonacular-recipe-food-nutrition-v1.p.rapidapi.com/recipes/searchComplex"

querystring = {"cuisine":origin,"type":"main course","minCalories":"150","maxCalories":"1500","minFat":"5","maxFat":"100","minProtein":"5","maxProtein":"100","minCarbs":"5","maxCarbs":"100","limitLicense":"false","offset":"0","number":number_of_recipe}

headers = {
    'x-rapidapi-host': "spoonacular-recipe-food-nutrition-v1.p.rapidapi.com",
    'x-rapidapi-key': rapid_api_key
    }

response = requests.request("GET", url, headers=headers, params=querystring).json()

print(response)
```

```python
response
```

```python
#list to store data
id = []
title = []
image = []
region = []
weightWatcherSmartPoints = []
readyInMinutes = []
sourceUrl = []
pricePerServing = []
healthScore = []
aggregateLikes = []
vegetarian = []
vegan = []
glutenFree = []
dairyFree = []
servings = []
instructions = []
summary = []
ingredients = []
calories = []
fat = []
carbs = []
protein = []
# sugar = []
# cholesterol = []
# sodium = []
# fiber = []
# iron = []
# vitamin_c = []
# vitamin_e = []
# vitamin_d = []
```

```python
for x in range(int(number_of_recipe)):
    id.append(response['results'][x]['id'])
    title.append(response['results'][x]['title'])
    image.append(response['results'][x]['image'])
    calories.append(response['results'][x]['calories'])
    protein.append(response['results'][x]['protein'])
    fat.append(response['results'][x]['fat'])
    carbs.append(response['results'][x]['carbs'])
    region.append(origin)
```

```python
#Get the ids of recipes and join them as a single string to pass on to next api call
list_of_ids = ','.join(map(str, id)) 
list_of_ids
```

```python
#Get the recipe summary with list of ids
url = "https://spoonacular-recipe-food-nutrition-v1.p.rapidapi.com/recipes/informationBulk"

querystring = {"includeNutrition":"true","ids":list_of_ids}

headers = {
    'x-rapidapi-host': "spoonacular-recipe-food-nutrition-v1.p.rapidapi.com",
    'x-rapidapi-key': rapid_api_key
    }

response2 = requests.request("GET", url, headers=headers, params=querystring).json()

print(response2)
```

```python
#Append api information received to list
for x in range(int(number_of_recipe)):
    weightWatcherSmartPoints.append(response2[x]['weightWatcherSmartPoints'])
    readyInMinutes.append(response2[x]['readyInMinutes'])
    sourceUrl.append(response2[x]['sourceUrl'])
    pricePerServing.append(response2[x]['pricePerServing'])
    healthScore.append(response2[x]['healthScore'])
    aggregateLikes.append(response2[x]['aggregateLikes'])
    vegetarian.append(response2[x]['vegetarian'])
    vegan.append(response2[x]['vegan'])
    glutenFree.append(response2[x]['glutenFree'])
    dairyFree.append(response2[x]['dairyFree'])
    servings.append(response2[x]['servings'])
    instructions.append(response2[x]['instructions'])
    summary.append(response2[x]['summary'])
```

```python
for x in range(int(number_of_recipe)):
    extendedIngredients = response2[x]['extendedIngredients']
    temp = []
    for y in range(len(extendedIngredients)):
        temp.append(response2[x]['extendedIngredients'][y]['originalString'])
    list_of_ingredients = '; '.join(map(str, temp))
    ingredients.append(list_of_ingredients)
```

```python
#make dictionary out of the lists
food_dict = {
    "id": id,
    "title": title,
    "image": image,
    "region": region,
    "weightWatcherSmartPoints": weightWatcherSmartPoints,
    "readyInMinutes": readyInMinutes,
    "sourceUrl": sourceUrl,
    "pricePerServing": pricePerServing,
    "healthScore": healthScore,
    "aggregateLikes": aggregateLikes,
    "vegetarian": vegetarian,
    "vegan": vegan,
    "glutenFree": glutenFree,
    "dairyFree": dairyFree,
    "servings": servings,
    "instructions": instructions,
    "summary": summary,
    "ingredients": ingredients,
    "calories": calories,
    "fat": fat,
    "carbs": carbs,
    "protein": protein,
#     "sugar": sugar,
#     "cholesterol": cholesterol,
#     "sodium": sodium,
#     "fiber": fiber,
#     "iron": iron,
#     "vitamin_c": vitamin_c,
#     "vitamin_e": vitamin_e,
#     "vitamin_d": vitamin_d
}
food_data = pd.DataFrame(food_dict)
food_data
```
