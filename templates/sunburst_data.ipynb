```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress
import copy
import json
from collections import defaultdict
```

```python
#Import merged food data.
merged_data = "Data/merged.csv"
merged_df = pd.read_csv(merged_data)
merged_df
```

```python
#filter columns for future use.
df = merged_df[['title', 'region', 'ingredients']]
instructions_df = merged_df[['title', 'region', 'instructions']]
```

```python
#Get unique region names.
region = df.drop_duplicates('region')
region_list = []
for i in range(len(region['region'])):
    region_list.append(region.iloc[i,1])
region_list
```

```python
#append region, dish name, and ingredients to food_dict as dictionary of list of dictionary of list of dictionary (format of sunburst chart)
food_dict = {} #make dictionary to hold
food_dict['name'] = 'food'
food_dict['children'] = []  #make a list in food_dict to hold dictionarys

for i in range(len(region_list)):
    region_dict = {} #make temp region dict to hold region name and the list of dish names.
    region_dict['name'] = f"{region_list[i]}"
    region_dict['children'] = [] #insert list into region_dict to hold dishes.
    food_dict['children'].append(region_dict) #append region into food_dict
    for x in range(len(df)):
        region_df = df[df['region'] == region_list[i]] #temp df that holds dataframe filtered by region
    list = [] #a list that will hold lists of ingredients.
    for row in region_df['ingredients']:
        splited=row.split('; ') #the ingredients are a string seperated by "; ", so I split them and append them into the temp list.
        list.append(splited)
    for y in range(len(region_df)):
        dish_dict = {} #a temp dish dictionary to hold dish name and ingredient list
        dish_dict['name'] = f"{region_df.iloc[y,0]}"
        dish_dict['children'] = [] #insert list into dish_dict to hold ingredients.
        region_dict['children'].append(dish_dict) #append dishes into region_dict
        ingredient_list = list[y] #loop through the list of lists of ingredients.
        for z in range(len(ingredient_list)):
            ingredient_dict = {} #a temp ingredient dictionary that will hold ingredient names and values.
            ingredient_dict['name'] = ingredient_list[z] #loop list of ingredients per dish
            ingredient_dict['value'] = 1 #set value of ingredients to 1 (format of sunburst chart)
            dish_dict['children'].append(ingredient_dict) #append ingredient_dict to dish_dict
```
